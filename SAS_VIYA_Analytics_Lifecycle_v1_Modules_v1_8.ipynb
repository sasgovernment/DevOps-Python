<B>LOAD to SAS Viya</B>

```python
#https://github.com/sassoftware/python-swat
import swat
import time
tic = time.perf_counter()
sess = swat.CAS('pdcesx12023.exnet.sas.com', 5570, 'sasdemo', 'Orion123')

file_name="cms_sdoh_wide_data3.csv"
indata="cms_sdoh_wide_data3"

sess.loadactionset(actionset="table")

#mf added conditional
if sess.table.tableExists(table=indata).exists:
    tbl = sess.dropTable(name=indata)

if not sess.table.tableExists(table=indata).exists:
    tbl = sess.upload_file(file_name, casout={"caslib":"casuser", "name":indata}, promote='True')
    
tbl.head()
sess.close()

toc = time.perf_counter()
print(f"Ran the code in {toc - tic:0.4f} seconds")
```
